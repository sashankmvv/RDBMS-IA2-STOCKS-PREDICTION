CALCULATING 5 DAY MOVING AVERAGE

Analysts can use the simple moving average as a signal to buy or sell. If a price movement breaks above the SMA, it could signal an opportunity to buy. If the moving average then begins to also trail up, this is used to confirm the buy decision. Conversely, if price movement starts to break below the SMA, it could be a signal to sign to sell. 

```python
n=0
sum=0
while(n!=5277):
    sum=0
    for i in range(n,n+5):
        sum=sum+df['Close'][i]
        
    df['5-DAY_SMA'][n+4]=sum/5
    n=n+1
```

```python
fig = px.line(df, x="Date", y=["Close"])
fig.add_scatter(x=df['Date'], y=df['5-DAY_SMA'],name='5-Day-SMA')
fig.show()
```

CALCULATING 14 DAY WEIGHTED MOVING AVERAGE


This is used  alongside to simple moving average. If the weighted average is lower relative to the simple average, it suggests that recent selling pressure could be more significant than some traders anticipate.

```python
n=0
sum=0
k=1
while(n!=5268):
    sum=0
    j=0
    for i in range(n,n+14):

        sum=sum+(j+1)*df['Close'][i]
        j=j+1

    df['14-DAY_WMA'][n+13]=sum/105
    n=n+1
```

```python
fig = px.line(df, x="Date", y=["Close"])
fig.add_scatter(x=df['Date'], y=df['14-DAY_WMA'],name='14-Day-WMA')
fig.show()
```

```python
df['D']=df['Close'].diff()
df['U']=df['Close'].diff()
```

```python
def positive_U(val):
    if val<0:
        return 0
    
    return val
def positive_d(val):
    if val>0:
        return 0
    
    return -1*val
```

```python
df['U']=df['U'].apply(positive_U)
df['D'] = df['D'].apply(positive_d)
```

```python
df['Relative_strength_index']=(100)-(100/(1+(df['U'].rolling(window=10).mean()/df['D'].rolling(window=10).mean())))
```
